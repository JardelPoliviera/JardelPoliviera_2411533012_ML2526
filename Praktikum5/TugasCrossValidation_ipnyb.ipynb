```python
import numpy as np
from sklearn.model_selection import KFold, cross_validate
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
```

```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
FOLDER_PATH = "/content/drive/MyDrive/titanic"

df_pertama = pd.read_csv(f"{FOLDER_PATH}/gender_submission.csv")
df_kedua = pd.read_csv(f"{FOLDER_PATH}/test.csv")
df_ketiga = pd.read_csv(f"{FOLDER_PATH}/train.csv")

df_pertama.head()
```

```python
features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']
X = df_ketiga[features].copy()
y = df_ketiga['Survived']

imputer = SimpleImputer(strategy='median')
X['Age'] = imputer.fit_transform(X[['Age']])
X['Fare'] = imputer.fit_transform(X[['Fare']])

encoder = LabelEncoder()
X['Sex'] = encoder.fit_transform(X['Sex'])
```

```python
model = RandomForestClassifier(random_state=42)
scoring_metrics = ['accuracy', 'precision', 'recall']
```

```python
k_values = [5, 10]

for k in k_values:
    kf = KFold(n_splits=k, shuffle=True, random_state=42)

    cv_results = cross_validate(model, X, y, cv=kf, scoring=scoring_metrics)

    print(f"=== Hasil Evaluasi K-Fold (k={k}) ===")
    print(f"Rata-rata Accuracy : {np.mean(cv_results['test_accuracy']):.4f}")
    print(f"Rata-rata Precision: {np.mean(cv_results['test_precision']):.4f}")
    print(f"Rata-rata Recall   : {np.mean(cv_results['test_recall']):.4f}")
    print("-" * 45)
```
